### Standardisation

Before we apply PCA we need to standardize/scale the data. For this we can use sklearn's preprocessing library

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(numerics)
```

```python
from sklearn.decomposition import PCA
```

```python
# Principal Component Analysis (PCA) is a method to reduce dimensions (from 2d to 3d)
# it tries to find the best possible subspace that explains most of the variance in the data
# how? it derives components from the original scaled features and plots the 2 most significant ones 
pca = PCA()
pca.fit(X_train)
```

```python
evr = pca.explained_variance_ratio_
evr
```

```python
fig = plt.figure(figsize=(10,8))
plt.plot(range(1, len(numerics.columns)+1), evr.cumsum(), marker='o', linestyle='--')
plt.xlabel('Number of Components', fontsize=18)
plt.ylabel('Cumulative Explained Variance',fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
for i, exp_var in enumerate(evr.cumsum()):
    if exp_var >= 0.8:
        n_comps = i + 1
        break
print("Number of components:", n_comps)
pca = PCA(n_components=n_comps)
pca.fit(X_train)
scores_pca = pca.transform(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = numerics

X_train, X_test = train_test_split(X, test_size=0.30, random_state=42) #excluded target variable here

scaler = StandardScaler()

scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test  = scaler.transform(X_test)

X_train = pd.DataFrame(X_train, columns = X.columns)
X_test  = pd.DataFrame(X_test , columns = X.columns)
```

# Performing PCA

```python
from sklearn.decomposition import PCA

pca = PCA(svd_solver='full')

pca.fit(X_train)

X_train = pca.transform(X_train)
X_test  = pca.transform(X_test)

cols = [ 'PCA_' + str(i) for i in list(range(1,len(X.columns)+1))]

X_train = pd.DataFrame(X_train, columns = cols)
X_test  = pd.DataFrame(X_test , columns = cols)

#X_train = pd.DataFrame(X_train, columns = X.columns)
#X_test  = pd.DataFrame(X_test , columns = X.columns)
```

# Getting Variances

```python
# Getting
variations = pca.explained_variance_ratio_*100 # Individual variances of each principal components
variations = np.round(variations, decimals=1)
variations
#print(pca.singular_values_)
```

```python
variances = pca.explained_variance_ratio_.tolist()
```

```python
variances[:2]
```

```python
total_variance = [sum(variances[:i]) for i in range(1, len(variances)+1)]
scree = pd.DataFrame({'num_pca': range(1,len(variances)+1),'variance': variances,'cum_variance': total_variance})
scree.head(31)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style("darkgrid", {"axes.facecolor": ".9"})

fig, ax = plt.subplots(1,2,figsize=(16,8))

sns.lineplot(x="num_pca", y="variance", data=scree, ax = ax[0], color = 'black')
ax[0].set_xticks(range(1,32))
sns.barplot(x="num_pca", y="cum_variance", data=scree, ax = ax[1], color = 'lightblue')
```

In this scenario it looks like we have a an elbow at 5, and on the left we see that with five principal components we are able to explain 80.5% of the variances (cf. table) 

Bear in mind that we only selected the numerical columns and we didn't take into account the categorical ones.
