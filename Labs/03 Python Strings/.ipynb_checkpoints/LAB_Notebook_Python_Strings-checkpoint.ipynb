In the cell below, find all the words that appear in the text and do not appear in the blacklist. You must parse the string but can choose any data structure you wish for the words that do not appear in the blacklist. Remove all non letter characters and convert all words to lower case.

```python
blacklist = ['and', 'as', 'an', 'a', 'the', 'in', 'it']

poem = """I was angry with my friend; 
I told my wrath, my wrath did end.
I was angry with my foe: 
I told it not, my wrath did grow. 

And I waterd it in fears,
Night & morning with my tears: 
And I sunned it with smiles,
And with soft deceitful wiles. 

And it grew both day and night. 
Till it bore an apple bright. 
And my foe beheld it shine,
And he knew that it was mine. 

And into my garden stole, 
When the night had veild the pole; 
In the morning glad I see; 
My foe outstretched beneath the tree."""

# Your code here:
clean_poem = poem.translate(str.maketrans(';', ':', string.punctuation))
clean_poem = clean_poem.lower()

clean_poem_split = clean_poem.split()

for word in clean_poem_split:
    for b in blacklist:
        if b == word:
            clean_poem_split.remove(word)
print("NOW IT'S ALL CLEAN':", clean_poem_split)

#Solution

poem.lower()

import re

#find items that dont belong in the list
regex = re.compile("XXXX")
lower_alphabet_str = regex.sub("", poem.lower())
lower_alphabet_str

unique_cleaned_words = set(lowered_alphabet_str.split(" "))
unique_cleaned_words

# finding items dont belong to the list 
white_listed_items = unique_cleaned_words - set(blacklist)
white_listed_items
```
