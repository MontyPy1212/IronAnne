```python
def state_group(i):
    # checking for NaN. If x is NaN, return x
    if not i == i:
        return i
    
    else:

        dct_ = {'Washington':'west_coast',
                'Arizona':'midwest',
                'Nevada': 'midwest',
                'California':'west_coast',
                'Oregon':'west_coast',
               }
        
        return dct_[i]

insurance_df_all_new ['region'] = insurance_df_all_new['state'].apply(state_group)
insurance_df_all_new[['region', 'state']]
```
